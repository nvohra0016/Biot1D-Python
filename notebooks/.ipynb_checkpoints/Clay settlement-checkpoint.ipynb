In this example we consider a physical scenario of soil consolidation and calculate the corresponding settlement. The example_number corresponding to this scenario is example_number = 5, 6, 7, or 8.

# Example 1

## example_number = 7

This example describes the consolidation of a column 10 [m] column of clay when an overburden stress of 0.1 [MPa] is applied at the top of the column. The simulation is run for 1 [year], i.e., 8760 [days].

Consider $\Omega = (0,10)$ [m], with $x = 0$ [m] being the top of the soil column. The source terms are 

$$f(x,t) = 0, g(x,t) = 0, \ \forall x \in \Omega, 0 < t < 8760.$$

The physical parameters used are

| Parameter | Value                    | Unit         |
|-----------|--------------------------|--------------|
| $E$       | 20                       | [MPa]        |
| $\nu$     | 0.30                     | [-]          |
| $\phi$    | 0.50                     | [-]          |
| $\beta_f$ | 4.16$\times 10^{-4}$     | [1/MPa]      |
| $\kappa$  | 1                        | [m$^2$]      |
| $\mu_f$   | 2.7822 $\times 10^{-13}$ | [MPa hr]     |
| $\rho_f$  | 998.21                   | [kg / m$^3$] |
| $\rho_s$  | 2700                     | [kg / m$^3$] |
| $G$       | 1.27290582 $\times 10^8$ | [m / hr$^2$] |


The boundary conditions are implemented as [1, 0, 0, 1] to allow for a free draining top and a fixed, impermeable bottom.
 
$$ \Tilde{\sigma}n = 10^{-1}, \ x = 0, u = 0, \ x = 10, $$


$$ p = 0, \ x = 0, q_f = 0, \ x = 10. $$

We run the simulation by dividing $\Omega$ into Nxcells = 100 spatial cells corresponding to a grid size $h = 0.1$[m] and a time step of $\tau = $ 8760/Ntcells = $ 2$ [hr].
