In this example, we test the accuracy of the code using manufactured solutions. The example numbers corresponding to manufactured solutions are example_number = 1, 2, or 3.

## Example 1
### example_number = 1

For $\Omega = (0,1)$, the analytical solution is given by

$$ u(x,t) = -\frac{1}{\pi} \cos{(\pi x)} \sin{\left(\frac{\pi t}{2} \right)}, \ p(x,t) =  \sin{(\pi x)} \sin{\left(\frac{\pi t}{2}\right)},  \ \forall x \in (0,1), \  0 < t < 1. $$

The source terms are given by

$$ f(x,t) = \left[ -(\lambda + 2\mu)\pi + \alpha \pi \right] \cos{(\pi x)} \sin{\left(\frac{\pi t}{2}\right)},$$

$$g(x,t) = \frac{\pi}{2} (\beta_f \phi + \alpha)\sin{(\pi x)}\cos{\left(\frac{\pi t}{2}\right)} + \frac{\kappa}{\mu}\pi^2 \sin{(\pi x)} \sin{\left(\frac{\pi t}{2}\right)}, \ \forall x \in (0,1), \ 0 < t < 1.$$

The physical parameters used are

| Parameter | Value    | Unit         |
|-----------|----------|--------------|
| $\lambda$ | 1        | [MPa]        |
| $\mu$     | 1        | [MPa]        |
| $\phi$    | 1        | [-]          |
| $\beta_f$ | 1        | [1/MPa]      |
| $\kappa$  | 1        | [m$^2$]      |
| $\mu_f$   | 1        | [MPa hr]     |
| $\rho_f$  | 1.296e23 | [kg / m$^3$] |
| $G$       | 0        | [m / hr$^2$] |

The boundary conditions for this simulation are flagged [1, 0, 0, 1] i.e., for displacement Neumann at $x = 0$ and Dirichlet at $x = 1$, and for pressure Dirichlet at $x = 0$ and Neumann at $x = 1$.

We run the simulation by dividing $\Omega$ into Nxcells = 20 spatial cells and a time step of $\tau = $ 1/Ntcells = $ 1/20$.
 

```python
import os
os.chdir('../')
%run driver.py driver 20 20 1
```
